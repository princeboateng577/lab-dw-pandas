### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

## Challenge 1: Understanding the data

In this challenge, you will use pandas to explore a given dataset. Your task is to gain a deep understanding of the data by analyzing its characteristics, dimensions, and statistical properties.

- Identify the dimensions of the dataset by determining the number of rows and columns it contains.
- Determine the data types of each column and evaluate whether they are appropriate for the nature of the variable. You should also provide suggestions for fixing any incorrect data types.
- Identify the number of unique values for each column and determine which columns appear to be categorical. You should also describe the unique values of each categorical column and the range of values for numerical columns, and give your insights.
- Compute summary statistics such as mean, median, mode, standard deviation, and quartiles to understand the central tendency and distribution of the data for numerical columns. You should also provide your conclusions based on these summary statistics.
- Compute summary statistics for categorical columns and providing your conclusions based on these statistics.

```python
# Your code here
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
insurance_company = pd.read_csv(url)
insurance_company
```

```python
print(insurance_company.dtypes)
#Customer, ST, GENDER, EDUCATION,Policy Type, Vehicle Class ... These are objects and they are appropriate since most of them are categorical data type while Customer is the only column for ID column
#Total Claim Ammount has the appropriate data type because it represnt monetary value
#Monthly Premium Auto has appropriate data type because it has monetary value
#Income represents numeric value and is appropriate
```

```python
# Check for any incorrect data types
for col in insurance_company.columns:
    if insurance_company[col].dtype == 'object':
        try:
            pd.to_numeric(insurance_company[col])
            print(f"Column '{col}' could potentially be converted to a numerical type.")
        except ValueError:
            pass  # Leave as object if it can't be converted
print()
```

```python
#Identify the number of unique values for each column
unique_values = insurance_company.nunique()
print("Number of unique values in each column:")
print(unique_values)
```

```python
# Determine categorical and numerical columns
categorical_columns = insurance_company.select_dtypes(include=['object']).columns
numerical_columns = insurance_company.select_dtypes(include=['number']).columns

print("Categorical columns and their unique values:")
for col in categorical_columns:
    print(f"{col}: {insurance_company[col].unique()}")

print("Numerical columns and their range of values:")
for col in numerical_columns:
    print(f"{col}: Min={insurance_company[col].min()}, Max={insurance_company[col].max()}")
```

```python
# Select numerical columns
numerical_columns = insurance_company.select_dtypes(include=['number'])

# Compute summary statistics
summary_stats = numerical_columns.describe().T  # Transpose for better readability
print("Basic Summary Statistics for Numerical Columns:")
print(summary_stats)

# Compute additional statistics
print("Additional Statistics:")
for col in numerical_columns.columns:
    print(f"Column: {col}")
    print(f"Mean: {numerical_columns[col].mean()}")
    print(f"Median: {numerical_columns[col].median()}")
    print(f"Mode: {numerical_columns[col].mode().iloc[0]}")  # Handle mode in case of multiple modes
    print(f"Standard Deviation: {numerical_columns[col].std()}")
    print(f"Quartiles (25%, 50%, 75%): {numerical_columns[col].quantile([0.25, 0.5, 0.75])}")
```

```python
# Select categorical columns
categorical_columns = insurance_company.select_dtypes(include=['object'])

# Compute summary statistics for categorical columns
print("Summary Statistics for Categorical Columns:")
for col in categorical_columns.columns:
    print(f"Column: {col}")
    print(f"Number of Unique Values: {categorical_columns[col].nunique()}")
    print(f"Unique Values: {categorical_columns[col].unique()}")
    print(f"Value Counts:{categorical_columns[col].value_counts()}")

#Conclusions:
#ST (State):

#Most customers are from CA (California), followed by WA (Washington).
#NV (Nevada) has the least number of customers.
#GENDER:

#The dataset is fairly balanced between male (M) and female (F) customers, though there are slightly more males.
#Policy Type:

#The majority of customers have a "Personal Auto" policy, suggesting the company primarily caters to individual customers.
#"Special Auto" policies are rare, which could indicate a niche product.

```

## Challenge 2: analyzing the data

### Exercise 1

The marketing team wants to know the top 5 less common customer locations. Create a pandas Series object that contains the customer locations and their frequencies, and then retrieve the top 5 less common locations in ascending order.

```python
# Your code here
# Create a Series object for customer locations and their frequencies
customer_locations = insurance_company['ST'].value_counts()

# Retrieve the top 5 less common locations in ascending order
top_5_less_common_locations = customer_locations.nsmallest(5)

# Display the result
print("Top 5 Less Common Customer Locations:")
print(top_5_less_common_locations)
```

### Exercise 2

The sales team wants to know the total number of policies sold for each type of policy. Create a pandas Series object that contains the policy types and their total number of policies sold, and then retrieve the policy type with the highest number of policies sold.

*Hint:*
- *Using value_counts() method simplifies this analysis.*
- *Futhermore, there is a method that returns the index of the maximum value in a column or row.*


```python
# Your code here
# Create a Series object for policy types and their total number of policies sold
policy_counts = insurance_company['Policy Type'].value_counts()

# Retrieve the policy type with the highest number of policies sold
most_sold_policy_type = policy_counts.idxmax()
most_sold_policy_count = policy_counts.max()

# Display the results
print("Total Policies Sold by Type:")
print(policy_counts)
print(f"Policy Type with the Highest Number of Policies Sold: {most_sold_policy_type} ({most_sold_policy_count} policies)")
```

### Exercise 3

The sales team wants to know if customers with Personal Auto have a lower income than those with Corporate Auto. How does the average income compare between the two policy types?

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
- Calculate the average income for each policy.
- Print the results.

```python
# Your code here
# Filter dataframes for each policy type
personal_auto_df = insurance_company.loc[insurance_company['Policy Type'] == 'Personal Auto']
corporate_auto_df = insurance_company.loc[insurance_company['Policy Type'] == 'Corporate Auto']

# Calculate the average income for each policy type
personal_auto_avg_income = personal_auto_df['Income'].mean()
corporate_auto_avg_income = corporate_auto_df['Income'].mean()

# Print the results
print(f"Average Income for Personal Auto: ${personal_auto_avg_income:.2f}")
print(f"Average Income for Corporate Auto: ${corporate_auto_avg_income:.2f}")
```

### Bonus: Exercise 4


Your goal is to identify customers with a high policy claim amount.

Instructions:

- Review again the statistics for total claim amount to gain an understanding of the data.
- To identify potential areas for improving customer retention and profitability, we want to focus on customers with a high policy claim amount. Consider customers with a high policy claim amount to be those in the top 25% of the total claim amount. Create a pandas DataFrame object that contains information about customers with a policy claim amount greater than the 75th percentile.
- Use DataFrame methods to calculate summary statistics about the high policy claim amount data. 

*Note: When analyzing data, we often want to focus on certain groups of values to gain insights. Percentiles are a useful tool to help us define these groups. A percentile is a measure that tells us what percentage of values in a dataset are below a certain value. For example, the 75th percentile represents the value below which 75% of the data falls. Similarly, the 25th percentile represents the value below which 25% of the data falls. When we talk about the top 25%, we are referring to the values that fall above the 75th percentile, which represent the top quarter of the data. On the other hand, when we talk about the bottom 25%, we are referring to the values that fall below the 25th percentile, which represent the bottom quarter of the data. By focusing on these groups, we can identify patterns and trends that may be useful for making decisions and taking action.*

*Hint: look for a method that gives you the percentile or quantile 0.75 and 0.25 for a Pandas Series.*

*Hint 2: check `Boolean selection according to the values of a single column` in https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9*

```python
# Your code here
# Review statistics for the Total Claim Amount
total_claim_stats = insurance_company['Total Claim Amount'].describe()
print("Summary Statistics for Total Claim Amount:")
print(total_claim_stats)

# Calculate the 75th percentile (top 25% cutoff)
claim_75th_percentile = insurance_company['Total Claim Amount'].quantile(0.75)

# Filter the DataFrame for customers with a claim amount above the 75th percentile
high_claim_customers = insurance_company[insurance_company['Total Claim Amount'] > claim_75th_percentile]

# Step 4: Calculate summary statistics for the high claim amount data
high_claim_summary = high_claim_customers.describe()

# Display the results
print("75th Percentile of Total Claim Amount:", claim_75th_percentile)
print("High Claim Customers DataFrame:")
print(high_claim_customers)

print("Summary Statistics for High Claim Customers:")
print(high_claim_summary)
```
